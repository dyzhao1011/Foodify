```python
# Creating a list of dictionaries where each dictionary is information about a fruit
fruitList = []
for fruit in response.json():
    fruitDict = {}
    fruitDict["Fruit"] = fruit.get("description")
    for nutrient in fruit.get("foodNutrients"):
        fruitDict[nutrient.get("nutrient").get("name")] = nutrient.get("amount")
       
    fruitList.append(fruitDict)
```

```python
# Create different verisons of the dataframe
# fruitDF1 : Replace NaN values with averages for each class
# fruitDF2 : Replace NaN values using KNN
# fruitDF3 : Replace NaN values with 0
fruitDF1 = fruitDF.copy()
grouped_fruits = fruitDF1.groupby('Type')

averages = grouped_fruits.mean()


for col in fruitDF.columns:
    if col != 'Type':  # Skip 'Class' column
        for fruit_type, group in grouped_fruits:
            avg = averages.loc[fruit_type, col]
            fruitDF1.loc[group.index, col] = group[col].fillna(avg)
            
X_1 = fruitDF1.iloc[:, :18]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
knn_imputer = KNNImputer(n_neighbors=5)
X_2 = knn_imputer.fit_transform(X)

fruitDF3 = fruitDF.copy()
fruitDF3.fillna(0, inplace=True)
X_3 = fruitDF3.iloc[:, :18]

X_2 = pd.DataFrame(X_2)
X_4 = X.copy()
X_list = [X_1, X_2, X_3, X_4]
```

```python
# Initialize different classifier models
# Comparing Separate models to each other (All data, all features)
dt = DecisionTreeClassifier(random_state = 11)
rf = RandomForestClassifier(random_state = 11)
svm = SVC(random_state = 11)
sgd = SGDClassifier(random_state = 11)
kn = KNeighborsClassifier()
ada = AdaBoostClassifier(random_state = 11)

classifiers = [dt, rf, svm, sgd, kn]
classifierNames = ['Decision\nTree', 'Random\nForest', 'SVM', 'SGD', 'KNN']
```

```python
# Comparing Separate models to each other (All data, all features)

modelAccuracies = {'Replace NaN with Averages': [], 'Replace NaN using KNN': [], 'Replace NaN with 0': [], 'Include NaN values': []}

for x, count in zip(X_list, range(0,4)):
    for classifier, name in zip(classifiers, classifierNames):

        # Perform cross-validation
        if name == "SVM" or name == "SGD" or name == "KNN" :
            scaler = StandardScaler()
            X_updated= scaler.fit_transform(x)
        else:
            X_updated = x
        if count == 3 and (name == "SVM" or name == "SGD" or name == "KNN"):
            cv_scores = 0
        else:
            sKFold = StratifiedKFold(n_splits=4, shuffle= True, random_state = 11)
            cv_scores = cross_val_score(classifier, X_updated, Y_encoded, cv=sKFold)
            

            # Print the cross-validation scores
        #print("Testing for ", name)
        #print("Mean CV Score:", np.mean(cv_scores), "\n")
        if count == 0:
            modelAccuracies['Replace NaN with Averages'].append(np.mean(cv_scores))
        elif count == 1:
            modelAccuracies['Replace NaN using KNN'].append(np.mean(cv_scores))
        elif count == 2:
            modelAccuracies['Replace NaN with 0'].append(np.mean(cv_scores))
        elif count == 3:
            modelAccuracies['Include NaN values'].append(np.mean(cv_scores))
            
```

```python
modelAccuracies['Replace NaN with Averages']
modelAccuracies['Replace NaN using KNN']
modelAccuracies['Replace NaN with 0']
modelAccuracies['Include NaN values']

x = np.arange(len(classifiers))
width = 0.2  # Width of the bars, adjust as needed

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - 1.5 * width, modelAccuracies['Replace NaN with Averages'], width, label='Replace NaN with Averages')
bars2 = ax.bar(x - 0.5 * width, modelAccuracies['Replace NaN using KNN'], width, label='Replace NaN using KNN')
bars3 = ax.bar(x + 0.5 * width, modelAccuracies['Replace NaN with 0'], width, label='Replace NaN with 0')
bars4 = ax.bar(x + 1.5 * width, modelAccuracies['Include NaN values'], width, label='Include NaN values')  # New bar


# Adding labels, title, and custom x-axis tick labels
ax.set_xlabel('Different Classifier Models', fontsize=12)
ax.set_ylabel('Accuracy', fontsize=12)
ax.set_title('Accuracy of Classifer Models using Different Methods of Handling Missing Data', fontsize=15)
ax.set_xticks(x)
ax.set_ylim(0, 1.0)
ax.set_xticklabels(classifierNames)
ax.legend()

for bars in [bars1, bars2, bars3, bars4]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{:.3f}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    fontsize=8,
                    ha='center', va='bottom')

plt.tight_layout()
plt.savefig("Different_Model_Accuracies_Handling_Missing_Data.png")
# Display the plot
plt.show()
```

```python
# Statistical analysis on the above probabilites
diff = [Averages - np.mean([KNN, zero], dtype='float64') for Averages, KNN, zero in zip(modelAccuracies['Replace NaN with Averages'], modelAccuracies['Replace NaN using KNN'], modelAccuracies['Replace NaN with 0'])]
avg_diff = sum(diff) / len(diff)
print('Performance of Averages over the other methods: ', avg_diff * 100)

rf_avg_diff = modelAccuracies['Replace NaN with Averages'][1] - np.mean([modelAccuracies['Replace NaN using KNN'][1], modelAccuracies['Replace NaN with 0'][1], modelAccuracies['Include NaN values'][1]], dtype='float64')
print('Performance of Averages over the other methods for Random Forest: ', rf_avg_diff * 100)

rf_avg_diff = modelAccuracies['Replace NaN with Averages'][0] - np.mean([modelAccuracies['Replace NaN using KNN'][0], modelAccuracies['Replace NaN with 0'][0], modelAccuracies['Include NaN values'][0]], dtype='float64')
print('Performance of Averages over the other methods for Decision Tree: ', rf_avg_diff * 100)
```
